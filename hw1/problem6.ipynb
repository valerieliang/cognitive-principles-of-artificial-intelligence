### A) Implement the cosine measure of word similarity

```python
## Write a function to compute the cosine similarity between two word vectors.

def cosine_similarity(x, y):
    return numpy.dot(x, y) / (numpy.linalg.norm(x) * numpy.linalg.norm(y))
```

### F) Write a function to perform the vector analogy task

```python
## Write a function that tests analogies.
def analogy_vector(w1, w2, w3, vectors):
    y = vectors[w1] - vectors[w2] + vectors[w3]
    best_word = None
    best_similarity = -1
    for word, vector in vectors.items():
        similarity = cosine_similarity(y, vector)
        if similarity > best_similarity:
            best_similarity = similarity
            best_word = word
    return best_word
```
